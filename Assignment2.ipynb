As shown above, if the file cannot be found by the interpreter or if it cannot be read as a text file, an exception will be raised.   
Modify function fileLength() so that a friendly message is printed instead:   
  
_fileLength('midterm.py')  
---> 358  
fileLength('idterm.py')  
---> File idterm.py not found._  

```python
def fileLength(fileName):
    try:
        infile = open(fileName)
        content = infile.read()
        infile.close()
        return len(content)
    except FileNotFoundError:
        return 'File {} not found'. format(fileName)    
    except UnicodeDecodeError as err:
        return 'Unable to decode/read file {}'.format(fileName)
    except OSError as err:
        return 'OS error - ({}) while accessing file {}'.format(repr(err), fileName)
    except Exception as err:
        return 'Error - ({}) occurred while accessing file - {}'.format(repr(err), fileName)

fileList = ['MidtermQuestion1.ipynb', 'MidtermQuestion.ipynb', 'Lectures\Lec 1\Week 1.pptx']   
for file in fileList:
    print('{} - {}'.format(file, fileLength(file)))
```

Write a class named Marsupial that can be used as shown below:  
  
_m = Marsupial()  
m.put_in_pouch('doll')  
m.put_in_pouch('firetruck')   
m.put_in_pouch('kitten')  
m.pouch_contents()  
--> ['doll', 'firetruck', 'kitten']_  

```python
class Marsupial:
    item = []
    def __init__(self):
        self.item = []
    def put_in_pouch(self, item):
        self.item.append(item)
    def pouch_contents(self):
        print(self.item)

m = Marsupial()
m.put_in_pouch("doll")
m.put_in_pouch("firetruck")
m.put_in_pouch("kitten")
m.pouch_contents()
```

Now write a class named Kangaroo as a subclass of Marsupial that inherits all the attributes of Marsupial and also:  
a. extends the Marsupial \_\_init\_\_ constructor to take, as input, the coordinates x and y of the Kangaroo object,  
b. supports method jump that takes number values dx and dy as input and moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and  
c. overloads the \_\_str\_\_ operator so it behaves as shown below.  
  
_k = Kangaroo(0,0)  
print(k) ---> I am a Kangaroo located at coordinates (0,0)  
k.put_in_pouch('doll')  
k.put_in_pouch('firetruck')

k.jump(1,0)

```python
class Kangaroo(Marsupial):
    x = 0
    y = 0
    def __init__(self, x, y):
        self.x = x
        self.y = y
        Marsupial.__init__(self);
    def jump(self, dx, dy):
        self.x += dx
        self.y += dy
    def __str__(self):        
        return f'I am a Kangaroo located at coordinates ({self.x}, {self.y})'
k = Kangaroo(0, 0)
print(k)

k.put_in_pouch("doll")
k.put_in_pouch("firetruck")
k.put_in_pouch("kitten")
k.pouch_contents()

k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)

k.jump(1, 3)
k.jump(0, 1)
k.jump(0, 2)
print(k)
```

#### Question 4
Write function collatz() that takes a positive integer x as input and prints the Collatz sequence starting at x.   
A Collatz sequence is obtained by repeatedly applying this rule to the previous number x in the sequence:  
  
x = &nbsp; 洧논/2&nbsp; 洧녰洧녭 洧논 洧녰洧 洧뉧롐洧뉧롐    
&nbsp; &nbsp; &nbsp; 3洧논+1 洧녰洧녭 洧논 洧녰洧 洧녶洧녬洧녬  
  
Your function should stop when the sequence gets to number 1.  
Your implementation must be recursive, without any loops.

```python
def collatz(n):
    print(int(n))
    if n == 1:
        return 1
    elif n%2 == 0:
        n /= 2
        return collatz(n)
    else:
        n = (3*n) + 1
        return collatz(n)

numList = [1, 10]
for num in numList:
    print(f"-----collatz({str(num)})-------")
    collatz(num)
```

#### Question 5

_binary(0) --> 0  
binary(1) --> 1  
binary(3) --> 11  
binary(9) --> 1001_

```python
bin = ""
def binary(n):  
    global bin
    bin = str(n%2)+bin 
    if n<=1:  
        print(bin)
        bin = ""
        return
    else:            
        return binary((int)(n/2))

deciList = [0, 1, 3, 9, 11, 12]
for deci in deciList:
    print(f'binary({str(deci)})', end = ' --> ')
    binary(deci)
```

#### Question 6
Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document.  
You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser.  
When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document.  
Each heading should be indented asfollows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc.  
Test your implementation using w3c.html.  
  
_infile = open('w3c.html')  
content = infile.read()  
infile.close()  
hp = HeadingParser()  
hp.feed(content)  
---> W3C Mission  
--->&nbsp;  Principles_

```python
from html.parser import HTMLParser
class HeadingParser(HTMLParser):
    def __init__(self):
    #Since Python 3, we need to call the __init__() function 
    #of the parent class
        super().__init__()
        self.reset()
        self.headerTag = ""
 
    #Defining what the methods should output when called by HTMLParser.
    def handle_starttag(self, tag, attrs):
        self.headerTag = ""
        # check for each whether tag's first char is header & tag's second char is numeric i.e h1-h6
        if tag[0].lower() == "h" and tag[1].isnumeric(): 
            # if yes, headerTag is set to tag
            self.headerTag = tag       
 
    def handle_data(self, data):
        #checks whether headerTag is set 
        if self.headerTag != "":
            data = data.strip() #removes extra space, if any                        
            if len(data) > 0: #prints if length of data is valid
                indent = int(self.headerTag[1]) - 1 #fetches numeric value from headerTag for indentation
                print(" "*indent + data)                 
 
infile = open('resource\w3c.html')
content = infile.read()
infile.close()
hp = HeadingParser()
hp.feed(content) 
```

#### Question 7
Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent.  
Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL.  
As shown below, indentation, specified by indent, should be used to indicate the depth of a URL.
  
_webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)  
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html  
&nbsp; http://reed.cs.depaul.edu/lperkovic/csc242/test2.html  
&nbsp; &nbsp; http://reed.cs.depaul.edu/lperkovic/csc242/test4.html  
&nbsp; http://reed.cs.depaul.edu/lperkovic/csc242/test3.html  
&nbsp; &nbsp; http://reed.cs.depaul.edu/lperkovic/csc242/test4.html_  

```python
from urllib.parse import urljoin
from urllib.request import urlopen
from html.parser import HTMLParser
class Collector(HTMLParser):
    #collects hyperlink URLs into a list'
    def __init__(self, url):
        #initializes parser, the url, and a list
        HTMLParser.__init__(self)
        self.url = url
        self.links = []
    
    def handle_starttag(self, tag, attrs):
        #collects hyperlink URLs in their absolute format
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    # construct absolute URL
                    absolute = urljoin(self.url, attr[1])
                    if absolute[:4] == 'http': # collect HTTP URLs                        
                        self.links.append(absolute)
                        
    #returns hyperlinks URLs in their absolute format                    
    def getLinks(self):        
        return self.links
    
urlList = []    
def webdir(url, depth, indent):        
    print((" "*indent) + url)
    if(depth <= 0):
        return
    
    # to avoid repetition of same URL
    if url in urlList:
        return
    else: 
        urlList.append(url)
    try:
        resource = urlopen(url)
        content = resource.read().decode()
        collector = Collector(url)
        collector.feed(content)
        for link in collector.getLinks():             
            webdir(link, depth-1, indent+1)
    except:
        print('Error at URL - ' + url)

webdir('https://reed.cs.depaul.edu/index.html', 2, 0)
```

#### Question 8
Write SQL queries on the below database table that return:  
a) All the temperature data.  
b) All the cities, but without repetition.  
c) All the records for India.  
d) All the Fall records.  
e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters.  
f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.  
g) The total annual rainfall for Cairo.  
h) The total rainfall for each season.  

```python
import sqlite3
def displayRecord(data):
    for a in data:
        print(str(a).ljust(15), end = ' | ')
    print()
con = sqlite3.connect('Assignment2_Q8.db')
try:    
    #Drop table Weather if already exists
    cur = con.cursor()
    cur.execute("DROP TABLE IF EXISTS Weather")    
    
    #create table weather
    cur.execute("CREATE TABLE Weather (City varchar(25), Country varchar(50), Season varchar(25), Temperature decimal(5,2), Rainfall decimal(5,2))")
    
    #open & read data from file
    infile = open('resource\Assignment2_SqlTable.csv')
    content = infile.read().split('\n')    
    infile.close()        
    
    # add file content in table weather
    for value in content[1::]:        
        cur.execute("INSERT INTO Weather VALUES (?, ?, ?, ?, ?)", value.split(','))    
    allColumns = ['City', 'Country', 'Season', 'Temperature', 'Rainfall']
    
    #view all records
    allRecord = cur.execute("select * from Weather")    
    print('All Record')
    displayRecord(allColumns)
    for record in allRecord:
        displayRecord(record)
    
    # a. all temperature
    allTemp = cur.execute("select Temperature from Weather")        
    print('\nAll Temperature')    
    for record in allTemp:
        displayRecord(record)
        
    # b. all cities without repetition
    allCities = cur.execute("select distinct(city) from Weather")    
    print('\nAll Cities without repetition')    
    for record in allCities:
        print(record)
        
    # c. All India records
    allIndia = cur.execute("select * from Weather where country = 'India'")    
    print('\nAll India records')
    displayRecord(allColumns)
    for record in allIndia:
        displayRecord(record)
    
    # d. All Fall records
    allFall = cur.execute("select * from Weather where season = 'Fall'")    
    print('\nAll Fall records')
    displayRecord(allColumns)
    for record in allFall:
        displayRecord(record)
        
    # e. Avg rainfall between 200mm - 400mm
    avgRain = cur.execute("select City, country, season, avg(rainfall) as AvgRainfall from Weather group by City, country, season having avg(rainfall) between 200 and 400")    
    print('\nAvg rainfall between 200mm - 400mm')
    displayRecord(['City', 'Country', 'Season', 'AvgRainfall'])
    for record in avgRain:
        displayRecord(record)
    
    # f. Avg fall temp above 20
    avgFallTemp = cur.execute("select City, country, avg(temperature) as AvgFallTemp from Weather where season = 'Fall' group by City, country having avg(temperature) >= 20 order by AvgFallTemp")    
    print('\nAvg fall temp above 20')
    displayRecord(['City', 'Country', 'AvgFallTemp'])
    for record in avgFallTemp:
        displayRecord(record)
    
    # g. Total annual rainfall in Cairo
    totalRainCairo = cur.execute("select sum(rainfall) from Weather where City = 'Cairo'")    
    print('\nTotal annual rainfall in Cairo ')
    for record in totalRainCairo:
        print(record)
    
    # h. Total rainfall for each season
    totalRainSeason = cur.execute("select season, (sum(rainfall)*100)/100.0 as TotalSeasonalRainfall from Weather group by season")    
    print('\nTotal rainfall for each season')
    displayRecord(['Season', 'TotalSeasonalRainfall'])
    for record in totalRainSeason:
        displayRecord(record)
    
    con.commit()
    con.close()    
except Exception as err:
    print('An error has occured. Error - {}'.format(err))    
    con.rollback()
    con.close()
```

Suppose list words is defined as follows: words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']    
Write list comprehension expressions that use list words and generate the following lists:  
  
a) _['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE', 'LAZY', 'DOG']_  
  
b) _['the', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']_  
  
c) _[3, 5, 5, 3, 5, 4, 3, 4, 3]_  
    &nbsp;&nbsp;(the list of lengths of words in list words).  
  
d) _[['THE', 'the', 3], ['QUICK', 'quick', 5], ['BROWN','brown', 5], ['FOX', 'fox', 3], ['JUMPS', 'jumps', 5], ['OVER', 'over', 4],['THE', 'the', 3], ['LAZY', 'lazy', 4], ['DOG', 'dog', 3]]_   
    &nbsp;&nbsp;(the list containing a list for every word of list words, where each list contains the word in uppercase and lowercase and the length of the word.)  
      
e) _['quick', 'brown', 'jumps', 'over', 'lazy']_   
    &nbsp;&nbsp;(the list of words in list words containing 4 or more characters.)

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over','the', 'lazy', 'dog']

# a. upper case
word_upper = [w.upper() for w in words]
print (word_upper)
print ()

# b. lower case
word_lower = [w.lower() for w in words]
print (word_lower)
print ()

# c. length of word in the word list
word_len = [len(w) for w in words]
print(word_len)
print ()

# d. list of uppercase, lowercase & length for each word
word_merge = [[word_upper[i], word_lower[i], word_len[i]] for i in range(len(words))]
print(word_merge)
print ()

# e. words conatining 4 or more chars
word_minLen4 = [w for w in words if len(w) >= 4]
print(word_minLen4)
```
